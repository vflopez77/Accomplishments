```python
import pandas as pd
import csv
import os
from sqlalchemy import create_engine
from config import pg_pw
from config import pg_user
```

```python
# ====== Connection ======
# Connecting to PostgreSQL by providing a sqlachemy engine
engine = create_engine('postgresql://'+pg_user+':'+pg_pw+'@127.0.0.1:5432/PH-EmployeeDB')
```

```python
# ====== Reading table ======
# Reading PostgreSQL table into a pandas DataFrame
retiring_titles_df = pd.read_sql('SELECT * FROM retiring_titles', engine)
retiring_titles_df
```

```python
# Write out DateFrame
csv_file = os.path.join("data","retiring_titles.csv")
retiring_titles_df.to_csv(csv_file)
```
