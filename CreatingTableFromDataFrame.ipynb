```python
import pandas as pd
from sqlalchemy import create_engine
from config import pg_pw
from config import pg_user
import os
```

```python
# ====== Connection ======
# Connecting to PostgreSQL by providing a sqlachemy engine
engine = create_engine('postgresql://'+'postgres:'+pg_pw+'@127.0.0.1:5432/Accomplishments')
```

```python
# Read raw tasks file and convert to dataframe
wt_tasks_csv = os.path.join("data","tasks.csv")
wt_tasks_df = pd.read_csv(wt_tasks_csv)
```

```python
# ====== Writing table ====== #
# Writing Dataframe to PostgreSQL and replacing table if it already exists
wt_tasks_df.to_sql(name='weighted_tasks', con=engine, if_exists = 'replace', index=False)
```

```python
# ====== Reading table ======
# Reading PostgreSQL table into a pandas DataFrame
data_df = pd.read_sql('SELECT * FROM weighted_tasks', engine)
data_df
```
